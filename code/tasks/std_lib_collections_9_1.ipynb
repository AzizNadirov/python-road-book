1. `namedtuple` istifadə edərək "Book" adlı struktur yaradın. Bu strukturda aşağıdakı sahələr olmalıdır: 
    - name - kitabın adı, 
    - author - müəllifin adı 
    - page_count - səhifə sayı
    
    Sonra 3 kitab yaradıb, çap edin.
   
   **Gözlənilən nəticə:**
   ```python
       Book(name='Koroğlu', author='Xalq yaradıcılığı', page_count=250)
       Book(name='Qarabağnamə', author='Mirzə Camal', page_count=180)
       Book(name='Dədə Qorqud', author='Xalq yaradıcılığı', page_count=300)
    ```


````python
from collections import namedtuple

Book = namedtuple('Book', ['name', 'author', 'page_count'])

print(Book(name='Koroğlu', author='Xalq yaradıcılığı', page_count=250))
print(Book(name='Qarabağnamə', author='Mirzə Camal', page_count=180))
print(Book(name='Dədə Qorqud', author='Xalq yaradıcılığı', page_count=300))
````

2. `Counter` istifadə edərək, `get_most_freq(seq)` funksiyasını yazın. Funksiya hər hansı ardıcıllıq alır və ardıcıllığın ən çox rast gəlinən elementini qaytarır.
    ```python
    def get_most_freq(seq):
        pass
        
    get_most_freq([1,1,1,2,3])    # 1
    ```


````python
from collections import Counter

def get_most_freq(seq):
    c = Counter(seq)
    return c.most_common(1)[0][0]

print(get_most_freq([1,1,1,2,3]))
````

3. Bank növbəsini simulyasiya edin. İlk öncə 3 müştəri növbəyə daxil olur: "Əli", "Vəli", "Səməd". Sonra "Leyla" növbəyə əlavə olunur. İlk müştəri xidmət alır (növbədən çıxır). Növbənin cari vəziyyətini göstərin.
   
   **Gözlənilən nəticə:**(sadəcə çap edin)
   ```
   İlkin növbə: deque(['Əli', 'Vəli', 'Səməd'])
   Leyla əlavə olundu: deque(['Əli', 'Vəli', 'Səməd', 'Leyla'])
   Əli xidmət aldı və çıxdı
   Cari növbə: deque(['Vəli', 'Səməd', 'Leyla'])
   ```


````python
from collections import deque

# ilkin növbə
que = deque(['Əli', 'Vəli', 'Səməd'])

# 'Leyla' əlavə olunur
que.append('Leyla')
print(que)

# Əli(soldan birinci) çıxır
gone = que.popleft()
print(f"Növbəni tərk edir: {gone}")

print(que)
````

4. "Contact" adlı namedtuple yaradın, sahələri: name, phone, email. `phone` və `email` susmaya görə dəyərləri boş sətr olsun. Üç ədəd kontakt yaradın:
   - Tam məlumatla, bütün sahələri doldurun.
   - Yalnız ad və telefonla

   **Gözlənilən nəticə:**

5. Verilmiş mətndə sözlərin tezliyini hesablayın və ən çox təkrarlanan 3 sözü tapın. Böyük-kiçik hərfləri eyniləşdirin və durğu işarələrini silin.
   
   **Giriş mətn:** "Python programming language is very powerful. Python is easy to learn. Python can do everything. Programming is the future."
   
   **Gözlənilən nəticə:**
```
   Total word count: 19
    Unique word count: 14 
    Top 3 most repeated words:
        'python' - 3 times
        'is' - 3 times
        'programming' - 2 times
   ```


````python
from collections import Counter

input_text = "Python programming language is very powerful. Python is easy to learn. Python can do everything. Programming is the future."
stripped_text = input_text.strip().lower().replace('.', '')

counter = Counter(
    stripped_text.split(' ')
)

n_unique = len(counter.keys())
# print(stripped_text.split(' '))
total_word_cnt = len(stripped_text.split(' '))
# print(f"Total word count: {total_word_cnt}")
# print(f"Unique words: {list(counter.keys())}")

top_3 = counter.most_common(3)

# printing:
print(f"Text: \n\t{input_text}\n")
print(f"Total word count: {total_word_cnt}")
print(f"Unique word count: {n_unique}")
print("Top 3 most repeated words:")
for top_item in top_3:
    print(f"\t'{top_item[0]}' - {top_item[1]} times")
````

6. Sadə kitabxana idarəetmə sistemi yaradın. Bu kitabxana sistemində:
    - Kitabxanada qeydiyyata alınmış kitablar və üzvlər var. Hər iki məvhum üçün adlı yazı yaratmaq olar.
    - Üzvlər kitabxanadan oxumaq üçün kitab icarələyə bilirlər. Deyək ki, bizim sistemdə hər kitabdan ancaq bir ədəd var. 
    - Kitab oxunmaya götürüləndə, bu kitab; oxumaqçün müraciət edənlər "gözləmə növbəsinə" daxil olur.
    - Kitab oxucu tərəfindən geri qaytarılan kimi, ilk gözləyən oxucuya verilir. Burada `deque` istifadə etmək olar.
    - Kitaba görə, onun kimdə - hansı üzvdə olduğunu bilmək olur.

Düşünün, sizcə bu cür sistemi necə dizayn etmək olar? Sinif şəklində reallaşdırmaq olarmı, hansı metod və xüsusiyyətlər olacaq?

````python
from collections import namedtuple, OrderedDict, deque


Book = namedtuple('Book', ['isbn', 'title', 'author'])
Member = namedtuple('Member', ['id', 'name', 'email'])


class Library:
    def __init__(self):
        self._books = {}  # isbn -> Book
        self._members = {}  # member_id -> Member
        self._borrowed = OrderedDict()  # isbn -> member_id (maintains borrow order)
        self._waitlists = {}  # isbn -> deque of member_ids
    
    def add_book(self, book: Book):
        """Add a book to the library"""
        if book.isbn not in self._books:
            self._books[book.isbn] = book
            self._waitlists[book.isbn] = deque()
    
    def add_member(self, member: Member):
        """Register a new library member"""
        self._members[member.id] = member
    
    def borrow_book(self, isbn: str, member_id: str) -> str:
        """
        Borrow a book if available, otherwise join waitlist.
        Returns status message.
        """
        if isbn not in self._books:
            return "Book not found in library"
        
        if member_id not in self._members:
            return "Member not registered"
        
        # Check if book is available
        if isbn not in self._borrowed:
            self._borrowed[isbn] = member_id
            return f"Book '{self._books[isbn].title}' borrowed successfully"
        
        # Book is taken, add to waitlist
        if member_id in self._waitlists[isbn]:
            return "You are already in the waitlist for this book"
        
        if self._borrowed[isbn] == member_id:
            return "You have already borrowed this book"
        
        self._waitlists[isbn].append(member_id)
        position = len(self._waitlists[isbn])
        return f"Book is currently borrowed. You are #{position} in the waitlist"
    
    def return_book(self, isbn: str, member_id: str) -> str:
        """
        Return a borrowed book and process waitlist.
        Returns status message.
        """
        if isbn not in self._borrowed:
            return "This book is not currently borrowed"
        
        if self._borrowed[isbn] != member_id:
            return "You haven't borrowed this book"
        
        # Remove from borrowed
        del self._borrowed[isbn]
        
        # Check waitlist
        if self._waitlists[isbn]:
            next_member_id = self._waitlists[isbn].popleft()
            self._borrowed[isbn] = next_member_id
            next_member = self._members[next_member_id]
            return f"Book returned. Automatically lent to {next_member.name} from waitlist"
        
        return "Book returned successfully"
    
    def get_waitlist_position(self, isbn: str, member_id: str) -> int | None:
        """Get position in waitlist (1-indexed), or None if not in waitlist"""
        if isbn not in self._waitlists:
            return None
        
        try:
            position = list(self._waitlists[isbn]).index(member_id) + 1
            return position
        except ValueError:
            return None
    
    def get_member_books(self, member_id: str) -> list[Book]:
        """Get all books currently borrowed by a member"""
        borrowed_books = []
        for isbn, borrower_id in self._borrowed.items():
            if borrower_id == member_id:
                borrowed_books.append(self._books[isbn])
        return borrowed_books
    
    def get_book_status(self, isbn: str) -> dict:
        """Get detailed status of a book"""
        if isbn not in self._books:
            return {"status": "not_found"}
        
        book = self._books[isbn]
        
        if isbn in self._borrowed:
            borrower_id = self._borrowed[isbn]
            borrower = self._members[borrower_id]
            
            return {
                "status": "borrowed",
                "book": book,
                "borrower": borrower.name,
                "waitlist_count": len(self._waitlists[isbn])
            }
        
        return {
            "status": "available",
            "book": book,
            "waitlist_count": len(self._waitlists[isbn])
        }
    
    def get_all_borrowed_books(self) -> list[tuple[Book, Member]]:
        """Get all currently borrowed books with their borrowers"""
        result = []
        for isbn, member_id in self._borrowed.items():
            book = self._books[isbn]
            member = self._members[member_id]
            result.append((book, member))
        return result
    
    def __str__(self):
        return (f"<Library: {len(self._books)} books, "
                f"{len(self._members)} members, "
                f"{len(self._borrowed)} borrowed>")


library = Library()

# Add books
book1 = Book("978-0-13-110362-7", "The C Programming Language", "Kernighan & Ritchie")
book2 = Book("978-0-13-468599-1", "Effective Python", "Brett Slatkin")
book3 = Book("978-0-13-235088-4", "Clean Code", "Robert Martin")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

# Add members
member1 = Member("M001", "Alice", "alice@email.com")
member2 = Member("M002", "Bob", "bob@email.com")
member3 = Member("M003", "Charlie", "charlie@email.com")

library.add_member(member1)
library.add_member(member2)
library.add_member(member3)

print("=== Library System Demo ===\n")

# Alice borrows C book
print(library.borrow_book("978-0-13-110362-7", "M001"))

# Bob tries to borrow the same book - joins waitlist
print(library.borrow_book("978-0-13-110362-7", "M002"))

# Charlie also joins waitlist
print(library.borrow_book("978-0-13-110362-7", "M003"))

# Check waitlist position
position = library.get_waitlist_position("978-0-13-110362-7", "M002")
print(f"\nBob's position in waitlist: #{position}")

# Check book status
print("\nBook status:")
status = library.get_book_status("978-0-13-110362-7")
print(f"  Status: {status['status']}")
print(f"  Borrowed by: {status['borrower']}")
print(f"  Waitlist: {status['waitlist_count']} people")

# Alice returns the book - automatically goes to Bob
print(f"\n{library.return_book('978-0-13-110362-7', 'M001')}")

# Check who has it now
status = library.get_book_status("978-0-13-110362-7")
print(f"Now borrowed by: {status['borrower']}")

# Check Charlie's new position
position = library.get_waitlist_position("978-0-13-110362-7", "M003")
print(f"Charlie's new position: #{position}")

# Check Alice's borrowed books
alice_books = library.get_member_books("M001")
print(f"\nAlice's borrowed books: {len(alice_books)}")

# Bob borrows another book
print(f"\n{library.borrow_book('978-0-13-468599-1', 'M002')}")

# Check all borrowed books
print("\nAll borrowed books:")
for book, member in library.get_all_borrowed_books():
    print(f"  - '{book.title}' by {member.name}")

print(f"\n{library}")
````
