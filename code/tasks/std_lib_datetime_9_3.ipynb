1.  Doğum tarixini alıb, ad gününə qədər qalan günlərin sayını qaytaracaq `days_until_birthday` funksiyasını yazın.

````python
from datetime import date

def days_until_birthday(birth_month: int, birth_day: int) -> int:
    """
    Calculate how many days until next birthday.
    
    Args:
        birth_month: Birth month (1-12)
        birth_day: Birth day (1-31)
        
    Returns:
        Number of days until next birthday
    """
    today = date.today()
    
    # bu il üçün doğum tarixi
    birthday_this_year = date(today.year, birth_month, birth_day)
    
    # Bu il üçün artıq ad günü olub keçibsə, gələn ili götürürük
    if birthday_this_year < today:
        birthday_next = date(today.year + 1, birth_month, birth_day)
    else:
        birthday_next = birthday_this_year
    
    # Fərqi tapırıq
    days_remaining = (birthday_next - today).days
    
    return days_remaining



print(f"Today is: {date.today()}")

# Example: 28 Yanvarda olsun
days = days_until_birthday(1, 28)
print(f"{days} days until your birthday!")
````

````python
from datetime import date

def calculate_age(birth_date: date) -> tuple[int, int, int]:
    """
    Calculate age in years, months, and days.
    
    Args:
        birth_date: Date of birth
        
    Returns:
        Tuple of (years, months, days)
    """
    today = date.today()
    
    # years
    years = today.year - birth_date.year
    
    # months
    months = today.month - birth_date.month
    
    # days
    days = today.day - birth_date.day
    
    # günlər və ya aylar mənfi alınırsa, düzənləyirik
    if days < 0:
        months -= 1
        if today.month == 1:
            prev_month = date(today.year - 1, 12, birth_date.day)
        else:
            prev_month = date(today.year, today.month - 1, birth_date.day)
        days = (today - prev_month).days
    
    if months < 0:
        years -= 1
        months += 12
    
    return (years, months, days)




birth = date(2000, 5, 15)
years, months, days = calculate_age(birth)
print(f"Age: {years} years, {months} months, {days} days")

# More tests
test_dates = [
    date(1990, 1, 1),
    date(2010, 12, 31),
    date(2005, 3, 27)
]

for birth_date in test_dates:
    y, m, d = calculate_age(birth_date)
    print(f"Born {birth_date}: {y} years, {m} months, {d} days old")
````

````python
from datetime import datetime

def parse_event_date(date_string: str) -> datetime:
    """
    Parse event date from various formats.
    
    Supported formats:
    - "2025-03-27" (ISO format)
    - "27/03/2025" (European format)
    - "03/27/2025" (US format)
    - "27.03.2025" (Dot separated)
    
    Args:
        date_string: Date as string
        format_hint: Optional hint like "ISO", "EU", "US", "DOT"
        
    Returns:
        datetime object
    """
    # Try to auto-detect format
    if "-" in date_string:
        # ISO format
        return datetime.strptime(date_string, "%Y-%m-%d")
    elif "/" in date_string:
        # Try European first (day/month/year)
        try:
            return datetime.strptime(date_string, "%d/%m/%Y")
        except ValueError:
            # If fails, try US format
            return datetime.strptime(date_string, "%m/%d/%Y")
    elif "." in date_string:
        # Dot separated
        return datetime.strptime(date_string, "%d.%m.%Y")
    else:
        raise ValueError(f"Unknown date format: {date_string}")



dates = [
    "2025-03-27",
    "27/03/2025",
    "03/27/2025",
    "27.03.2025"
]

print("Parsing different date formats:")
for date_str in dates:
    parsed = parse_event_date(date_str)
    print(f"{date_str} -> {parsed}")
````

4. Beynəlxalq görüş(zəng) çeviricisi yazın. Görüşün zamanı(`datetime`) və saat qurşağı var. Çevirici görüşün zamanını, saat qurşağını və iştirakçıların saat qurşaqlarından ibarət siyahı alır. Məqsəd - görüşün siyahıdakı hər bir saat qurşağı üçün saatını tapmaqdır. Məsələn, görüş Bakı vaxtı ilə saat 19-00 üçün (tz: Asia/Baku) təyin olunubsa, Asia/Tokyo(UTC+9) saat qurşağı üçün bu saat 10-00 olacaq.
```python
def schedule_meeting(
    meeting_time: datetime,
    meeting_timezone: str,
    participant_timezones: list[str]
) -> dict[str, datetime]:
    """
    Convert meeting time to all participant time zones.
    
    Args:
        meeting_time: Meeting datetime (naive)
        meeting_timezone: Timezone where meeting is scheduled (e.g., "Asia/Baku")
        participant_timezones: List of participant timezones
        
    Returns:
        Dictionary mapping timezone: local meeting time
        
    Example:
        Meeting: 2025-03-27 14:00 in Baku (Asia/Baku)
        Participants: ["Europe/London", "America/New_York", "Asia/Tokyo"]
        
        Returns:
        {
            "Asia/Baku": 2025-03-27 14:00:00+04:00,
            "Europe/London": 2025-03-27 10:00:00+00:00,
            "America/New_York": 2025-03-27 06:00:00-04:00,
            "Asia/Tokyo": 2025-03-27 19:00:00+09:00
        }
    """
    pass
```


````python
from datetime import datetime
from zoneinfo import ZoneInfo

def schedule_meeting(
    meeting_time: datetime,
    meeting_timezone: str,
    participant_timezones: list[str]
) -> dict[str, datetime]:
    """
    Convert meeting time to all participant time zones.
    
    Args:
        meeting_time: Meeting datetime (naive)
        meeting_timezone: Timezone where meeting is scheduled
        participant_timezones: List of participant timezones
        
    Returns:
        Dictionary mapping timezone -> local meeting time
    """
    meeting_tz = ZoneInfo(meeting_timezone)
    aware_meeting_time = meeting_time.replace(tzinfo=meeting_tz)
    
    result = {}
    result[meeting_timezone] = aware_meeting_time
    
    for tz in participant_timezones:
        if tz != meeting_timezone:
            participant_tz = ZoneInfo(tz)
            local_time = aware_meeting_time.astimezone(participant_tz)
            result[tz] = local_time
    
    return result



print("=== Meeting Scheduler ===\n")
meeting = datetime(2025, 3, 27, 14, 0, 0).astimezone(ZoneInfo("Asia/Baku"))

participants = ["Asia/Baku", 
                "Europe/London", 
                "America/New_York", 
                "Asia/Tokyo"]

print("Meeting scheduled for 14:00 in Baku")
print("\nLocal times for participants:")
times = schedule_meeting(meeting, "Asia/Baku", participants)

for tz, local_time in times.items():
    print(f"  {tz:20} -> {local_time.strftime('%Y-%m-%d %H:%M %Z')}")
````
