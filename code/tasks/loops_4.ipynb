1. Tələbələr siyahısı velib: `["Rəşad", "Əziz", "Ülvi", "Hikmət"]`. Hər bir tələbənin adını çap edin. 

````python
students = ["Rəşad", "Əziz", "Ülvi", "Hikmət"]

# for dövrü ilə:
for name in students:
    print(name)
````

---

````python
# while ilə də etmək olar, amma daha qəliz olacaq:
i = len(students)

while i:
    print(students[len(students)-i])
    i -= 1
    
````

2. Bu dəfə hər bir tələbənin adını və sait və ya samitlə başlamasını çap edin.

````python
sait = "aıoueəiöü"
for name in students:
    if name[0].lower() in sait:
        label = "sait"
    else: 
        label = "samit"
    
    print(f"{name} - {label}")
````

---

3. Siyahı generatorlarından istifadə edərək, ancaq samitlə başlayan tələbələrdən ibarət olan `students_consonants` siyahısını yaradın.

````python
students_consonants = [student for student in students if student.lower()[0] in sait]
print(students_consonants)
# ya da:

students_consonants = [student for student in students if student.lower().startswith(tuple(sait))]
print(students_consonants)
````

4. Siyahı generatorlarından istifadə edərək, tələbələr üçün email ünvanları siyahısı yaradın. Email domeni qismində `"@school.edu"` istifadə edin. 
Adları standartlaşdırmaq üçün adlardakı hərfləri bu cür əvəzləyin: `ə-e, ü-u, ö-o, ş-sh, ç-ch, ğ-gh, ı-i`

````python
print('students: ', students)

domain = "school.edu"
replacements = {
    'ə': 'e', 
    'ü': 'u', 
    'ö': 'o', 
    'ş': 'sh', 
    'ç': 'ch', 
    'ğ': 'gh', 
    'ı': 'i'
}

emails = [f"{student_name}@{domain}".lower() for student_name in students]  # ilkin email siyahısı
emails_clean = []
for email in emails:
    print(f"Cleaning email: {email}")
    email_cln = email
    for repl in replacements.keys():
        email_cln = email_cln.replace(repl, replacements[repl])

    print(f"Clean email: {email_cln}")
    emails_clean.append(email_cln)
    print('----------------------------------------')
print("Clean Emails: ", emails_clean)
````

````python
# və ya bir sətrə yazmaq olar.
[''.join(replacements.get(c, c) for c in f"{student}@{domain}".lower()) for student in students]
````

5. Tələbələrin alınmış adlar və email ünvanlarından istifadə etməklə, tələbənin ad və email ünvanlarını saxlayacaq lüğətlər siyahısını hazırlayın. Siyahı generatorundan və `zip` funksiyasından istifadə edə bilərsiniz.

````python
##!  Fərz edək ki, tələbə adları və email ardıcıllıqları eynidir.

print('students: ', students)
print('emails: ', emails_clean)

student_email_pairs = [{student_name: student_email} for student_name, student_email in zip(students, emails_clean)]

print(student_email_pairs)
````
