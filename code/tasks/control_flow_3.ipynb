1. İstifadəçi `input` funksiyası ilə ədəd daxil etməlidir. Ədəd 10-a tam bölünənə olarsa `"Onluq!"`, 5-ə bölünən olarsa `"Beşlik!"`, 3-ə bölünən olarsa `"Üçlük"`, 2 üçün `"Cütlük!"`, əks halda isə `"Nəsə başqa bir şey!"` çap edin.

```python
n = int(input("Number: "))
print(f"n: {n}")

if n % 10 == 0: # 10-a tam bölünür
    print(f"Onluq!")
elif n % 5 == 0: # 5-ə tam bölünür
    print("Beşlik!")

elif n % 3 == 0: # 3-ə tam bölünür
    print("Üçlük!")
    
elif n % 2 == 0: # 2-üçün
    print("Cütlük!")
    
else: # əks halda
    print("Nəsə başqa bir şey!")
```

2. istifadəçi iki ədəd daxil edir. Daha sonra istifadəçi `+`, `-`, `*` və ya `/` daxil edir. Daxil olunan əməliyyata uyğun olaraq ədədlərin cəmini, fərqini, hasilini və ya qismətini çap edin.

```python
a = int(input("a: "))
b = int(input("b: "))

ops = ['+', '-', '*', '/']  # bütün mümkün əməliyyatlar
operation = input(f"Operation {ops}: ")

print(f"a: {a}; b: {b}; operation: {operation}")

if operation not in ops:
    print(f"Incorrect Operation: '{operation}'. Correct ones: {ops}")

else:
    if operation == '+':
        print(a+b)
    elif operation == '-':
        print(a-b)
    elif operation == '*':
        print(a * b)
        
    elif operation == '/':
        print(a/b)

```

3. Daxil olunan sətrin polindrom olub-olmamasını çap edən proqram yazın. Polindrom  sözlər, hər iki istiqamətdə  - soldan-sağa və sağdan-sola eyni cür yazılan sözlərdir: ata, qapaq, mum və sairə. 

```python
# girişin sətr olduğunu nəzərə alsaq, əgər `sətr == sətrin_tərsi` bizə True qaytararsa, deməli sətr polindromdur. 
# Sətrin tərsini isə dilimləmənin addım hissəsini -1 kimi təyin etməklə almaq olar

string = "word"

print(string[: : -1])   # drow, tərsini verir

if string == string[::-1]:
    print("String is polindrime")
else:
    print("String is not polindrime")
    
    
## İndi isə polindrome ilə yoxlayaq:

string = "qapaq"

print(string[: : -1])

if string == string[::-1]:
    print("String is polindrime")
else:
    print("String is not polindrime")
```
