1. Bir `ascii_az(string)` funksiyası yaradın. Funksiya, Azərbaycan əlifbası hərflərini - onların ingilis dili qarşılığı ilə əvəz edəcək. Əvəzləmə üçün cütlər: `ə-e, ü-u, ö-o, ş-sh, ç-ch, ğ-gh, ı-i` 

```python
def ascii_az(string: str) -> str:
    # ...
        
ascii_az("Python öyrənirəm!")
#Funksiya qaytaracaq: "Python oyrenirem!"
```

````python
def ascii_az(string: str)->str:
    """
    ascii_az(string: str) -> str 
    Replaces Azerbaijan letters with their english equivilants.
    """
    replacements = {
        'ə': 'e',
        'ğ': 'gh',
        'ç': 'ch',
        'ı': 'i',
        'ü': 'u',
        # böyükləri üçün
        'Ə': 'E',
        'Ğ': 'Gh',
        'Ç': 'Ch',
        'I': 'I',
        'Ü': 'U',
    }
    for k, v in replacements.items():
        string = string.replace(k, v)
    return string

ascii_az("Python öyrənirəm!")
````

4. Fibonaçi ardıcıllığının n-ci həddini qaytaran funksiya yazın. Fibonaçi ardıcıllığında hər bir element, özündən öncəki iki elementin cəminə bərabərdir. Ardıcıllığın ilk 7(0 da sayılmasa 6) elementi belədir: `[0,1,1,2,3,5,8]`. 
```python
def fibonacci(n: int)->int:
    # Base cases: fib(0) = 0, fib(1) = 1
    # Recursive case: fib(n) = fib(n-1) + fib(n-2)
    pass

fibonacci(6)   # 8
```

````python
def fibonacci(n: int)->int:
    """Returns the nth number of the fibonacci sequence"""
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
fibonacci(6)
````

Funksiya geriyə sayğac funksiyası qaytarmalıdır, bu funksiyanı hər dəfə çağırdıqda öncəki çağırışdan bir vahid böyük ədəd qaytarmalıdır. `create_counter` fabrik funksiyası `start` arqumentini almalıdır, bu arqument sayğacın neçədən başlamalı olduğunu təyin edəcək, susmaya görə dəyəri `0` olmalıdır.
```python
def create_counter(start: int=0):
```

````python
def create_counter(start: int=0)->callable:
    """
    Returns a counter function
    """
    def counter():
        nonlocal start
        start += 1
        return start
    return counter

counter1 = create_counter()
counter2 = create_counter(10)

print(counter1())  # Should output: 1
print(counter1())  # Should output: 2
print(counter2())  # Should output: 11
print(counter1())  # Should output: 3
print(counter2())  # Should output: 12
````

6. `counter` adlı dekorator funksiya yaradın. Dekorator, bürünən funksiyanın hər çağırılışı zamanı, neçənci dəfə çağrıldığını çap etməlidir.
```python
def counter(func):
    # Your implementation here
    pass

@counter
def greet(name):
    return f"Hello, {name}!"

greet("Ali") # greet function called 1 time(s)
greet("Ali") # greet function called 2 time(s)
greet("Ali") # greet function called 3 time(s)    
```

````python
from functools import wraps

def counter(func: callable)->callable:
    count = 0
    @wraps(func)
    def wrapper(*args, **kwargs):
        nonlocal count
        res = func(*args, **kwargs)
        count += 1
        print(f"{func.__name__} called {count} times.")
        return res
    
    return wrapper


@counter
def greet(name):
    return f"Hello, {name}!"

greet("Ali") # greet function called 1 time(s)
greet("Ali") # greet function called 2 time(s)
greet("Ali") # greet function called 3 time(s)  
````

```python
def custom_range(start, stop, step=1):
    """
    Custom range generator function
    
    Args:
        start: Starting value (inclusive)
        stop: Ending value (exclusive)
        step: Step size (default: 1)
    """
    # Funksiyanın generator olması üçün yield istifadə edin.
    # `step` parametri həm mənfi, həm də müsbət ola bilər.
    pass

print(custom_range(0, 4))
# <generator object custom_range at ...>
print(list(custom_range(0, 4)))
# [0, 1, 2, 3]
```

````python

# while ilə
def custom_range(start: int, stop: int, step: int=1):
    while start < stop:
        yield start
        start += step

# rekursiv formada
def custom_range_recursive(start: int, stop: int, step: int=1):
    if start < stop:
        yield start
        yield from custom_range_recursive(start + step, stop, step)

# ən bəd ayaqda, for ilə:
def custom_range_for(start: int, stop: int, step: int=1):
    for i in range((stop - start + step - 1) // step):
        yield start + i * step


print(custom_range(0, 4))
print("---------------------------------------")
print(list(custom_range(0, 4)))
print("---------------------------------------")
print(list(custom_range_recursive(0, 4)))
print("---------------------------------------")
print(list(custom_range_for(0, 4)))
````
