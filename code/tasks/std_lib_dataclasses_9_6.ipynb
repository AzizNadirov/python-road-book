````python
from dataclasses import dataclass

@dataclass
class Student:
    """
    Student model with grades.
    
    Students are compared by average_grade first, then by name.
    """
    name: str
    surname: str
    grades: list[int]  # List of grades (0-100)
    
    def calculate_average(self) -> float:
        """Calculate average grade"""
        if not self.grades:
            return 0.0
        return sum(self.grades) / len(self.grades)
    
    def get_letter_grade(self) -> str:
        """
        Convert average to letter grade.
        
        90-100: A
        80-89: B
        70-79: C
        60-69: D
        0-59: F
        """
        avg = self.calculate_average()
        if avg >= 90:
            return 'A'
        elif avg >= 80:
            return 'B'
        elif avg >= 70:
            return 'C'
        elif avg >= 60:
            return 'D'
        else:
            return 'F'
    
    def __lt__(self, other):
        """Less than comparison for sorting"""
        # Compare by average first
        if self.calculate_average() != other.calculate_average():
            return self.calculate_average() < other.calculate_average()
        # If averages are equal, compare by name
        return self.name < other.name
    
    def __le__(self, other):
        return self < other or self == other
    
    def __gt__(self, other):
        return not self <= other
    
    def __ge__(self, other):
        return not self < other


print("=== Student Grade System ===\n")

students = [
    Student("Ali", "Aliyev", [85, 90, 88, 92]),
    Student("Aysel", "Mammadova", [95, 98, 93, 96]),
    Student("Kamran", "Hasanov", [70, 75, 72, 68]),
    Student("Leyla", "Ismayilova", [85, 90, 88, 92])  # Same avg as Ali
]

# Print students
print("All students:")
for student in students:
    avg = student.calculate_average()
    letter = student.get_letter_grade()
    print(f"{student.name} {student.surname}: {avg:.2f} ({letter})")

# Sort students by grade (highest first)
sorted_students = sorted(students, reverse=True)
print("\n=== Top Students ===")
for i, student in enumerate(sorted_students, 1):
    avg = student.calculate_average()
    letter = student.get_letter_grade()
    print(f"{i}. {student.name} {student.surname}: {avg:.2f} ({letter})")
````

````python
from dataclasses import dataclass, field

@dataclass(frozen=True)
class Product:
    """
    Immutable product information.
    """
    id: int
    name: str
    price: float
    category: str  # "Electronics", "Food", "Clothing", etc.
    
    def __post_init__(self):
        """Validate product data"""
        if self.price <= 0:
            raise ValueError(f"Price must be positive, got {self.price}")
        if not self.name or not self.name.strip():
            raise ValueError("Product name cannot be empty")

@dataclass
class CartItem:
    """
    Item in shopping cart with quantity.
    """
    product: Product
    quantity: int = 1
    
    def get_total(self) -> float:
        """Calculate total price for this item"""
        return self.product.price * self.quantity
    
    def __post_init__(self):
        """Validate cart item"""
        if self.quantity <= 0:
            raise ValueError(f"Quantity must be positive, got {self.quantity}")

@dataclass
class ShoppingCart:
    """
    Shopping cart with items and discount.
    """
    items: list[CartItem] = field(default_factory=list)
    discount_percent: float = 0.0  # 0-100
    
    def add_item(self, product: Product, quantity: int = 1):
        """
        Add product to cart.
        If product already exists, increase quantity.
        """
        # Check if product already in cart
        for item in self.items:
            if item.product.id == product.id:
                # Update quantity
                item.quantity += quantity
                return
        
        # Product not in cart, add new item
        self.items.append(CartItem(product, quantity))
    
    def remove_item(self, product_id: int):
        """Remove item from cart by product ID"""
        self.items = [item for item in self.items if item.product.id != product_id]
    
    def get_subtotal(self) -> float:
        """Calculate total before discount"""
        return sum(item.get_total() for item in self.items)
    
    def get_discount_amount(self) -> float:
        """Calculate discount amount"""
        return self.get_subtotal() * (self.discount_percent / 100)
    
    def get_total(self) -> float:
        """Calculate final total after discount"""
        return self.get_subtotal() - self.get_discount_amount()
    
    def get_items_by_category(self, category: str) -> list[CartItem]:
        """Get all items in a specific category"""
        return [item for item in self.items if item.product.category == category]


print("=== Shopping Cart Demo ===\n")

# Create products
laptop = Product(1, "Laptop", 1200.0, "Electronics")
mouse = Product(2, "Mouse", 25.0, "Electronics")
apple = Product(3, "Apple", 2.5, "Food")

# Create cart
cart = ShoppingCart()

# Add items
print("Adding items to cart...")
cart.add_item(laptop, 1)
cart.add_item(mouse, 2)
cart.add_item(apple, 5)

# Add same product again (should increase quantity)
print("Adding more mice...")
cart.add_item(mouse, 1)  # Now should have 3 mice

print(f"\nSubtotal: ${cart.get_subtotal():.2f}")

# Apply 10% discount
cart.discount_percent = 10
print(f"Discount (10%): ${cart.get_discount_amount():.2f}")
print(f"Total: ${cart.get_total():.2f}")

# Show electronics
print("\nElectronics in cart:")
for item in cart.get_items_by_category("Electronics"):
    print(f"  {item.product.name} x{item.quantity}: ${item.get_total():.2f}")

# Show all items
print("\nAll items in cart:")
for item in cart.items:
    print(f"  {item.product.name} x{item.quantity} @ ${item.product.price} = ${item.get_total():.2f}")

# Test validation
print("\n=== Testing Validation ===")
try:
    bad_product = Product(4, "", -10.0, "Invalid")
except ValueError as e:
    print(f"Caught error: {e}")

try:
    bad_item = CartItem(laptop, -5)
except ValueError as e:
    print(f"Caught error: {e}")
````

````python
from dataclasses import dataclass, field, asdict
from datetime import datetime, timedelta

@dataclass(frozen=True)
class Book:
    """Immutable book information"""
    isbn: str
    title: str
    author: str
    year: int
    
    def __post_init__(self):
        """Validate book data"""
        if len(self.isbn) != 13:
            raise ValueError(f"ISBN must be 13 characters, got {len(self.isbn)}")
        if not self.title.strip():
            raise ValueError("Title cannot be empty")
        if not self.author.strip():
            raise ValueError("Author cannot be empty")
        current_year = datetime.now().year
        if not (1000 <= self.year <= current_year):
            raise ValueError(f"Year must be between 1000 and {current_year}")

@dataclass(frozen=True)
class Member:
    """Immutable library member"""
    id: str
    name: str
    email: str
    join_date: datetime = field(default_factory=datetime.now)
    
    def __post_init__(self):
        """Validate member data"""
        if not self.id.strip():
            raise ValueError("Member ID cannot be empty")
        if '@' not in self.email:
            raise ValueError("Invalid email address")

@dataclass
class Loan:
    """Book loan record"""
    book: Book
    member: Member
    loan_date: datetime = field(default_factory=datetime.now)
    due_date: datetime = field(default=None)
    return_date: datetime = field(default=None)
    
    def __post_init__(self):
        """Set due date if not provided (14 days from loan date)"""
        if self.due_date is None:
            # For frozen dataclass, use object.__setattr__
            object.__setattr__(self, 'due_date', self.loan_date + timedelta(days=14))
    
    def is_overdue(self) -> bool:
        """Check if loan is overdue"""
        if self.return_date is not None:
            return False  # Already returned
        return datetime.now() > self.due_date
    
    def days_overdue(self) -> int:
        """Calculate days overdue (0 if not overdue)"""
        if not self.is_overdue():
            return 0
        return (datetime.now() - self.due_date).days
    
    def return_book(self):
        """Mark book as returned"""
        object.__setattr__(self, 'return_date', datetime.now())

@dataclass
class Library:
    """Library management system"""
    name: str
    books: list[Book] = field(default_factory=list)
    members: list[Member] = field(default_factory=list)
    loans: list[Loan] = field(default_factory=list)
    
    def add_book(self, book: Book):
        """Add book to library"""
        self.books.append(book)
    
    def add_member(self, member: Member):
        """Register new member"""
        self.members.append(member)
    
    def loan_book(self, isbn: str, member_id: str) -> Loan:
        """
        Loan a book to a member.
        
        Raises:
            ValueError: If book not found, member not found, or book already loaned
        """
        # Find book
        book = None
        for b in self.books:
            if b.isbn == isbn:
                book = b
                break
        
        if book is None:
            raise ValueError(f"Book with ISBN {isbn} not found")
        
        # Find member
        member = None
        for m in self.members:
            if m.id == member_id:
                member = m
                break
        
        if member is None:
            raise ValueError(f"Member {member_id} not found")
        
        # Check if book is already on loan
        for loan in self.loans:
            if loan.book.isbn == isbn and loan.return_date is None:
                raise ValueError(f"Book '{book.title}' is already on loan")
        
        # Create loan
        loan = Loan(book, member)
        self.loans.append(loan)
        return loan
    
    def return_book(self, isbn: str):
        """Return a borrowed book"""
        for loan in self.loans:
            if loan.book.isbn == isbn and loan.return_date is None:
                loan.return_book()
                return
        raise ValueError(f"No active loan found for ISBN {isbn}")
    
    def get_overdue_loans(self) -> list[Loan]:
        """Get all overdue loans"""
        return [loan for loan in self.loans if loan.is_overdue()]
    
    def get_member_loans(self, member_id: str) -> list[Loan]:
        """Get all loans for a member"""
        return [loan for loan in self.loans if loan.member.id == member_id]
    
    def export_to_dict(self) -> dict:
        """Export library data to dictionary"""
        return asdict(self)





print("=== Library Management System ===\n")

library = Library("City Library")

# Add books
book1 = Book("9780140328721", "1984", "George Orwell", 1949)
book2 = Book("9780451524935", "Animal Farm", "George Orwell", 1945)
library.add_book(book1)
library.add_book(book2)
print(f"Added {len(library.books)} books to library")

# Add members
member1 = Member("M001", "Ali Aliyev", "ali@email.com")
library.add_member(member1)
print(f"Registered {len(library.members)} members")

# Loan book
print("\n=== Loan Transaction ===")
loan = library.loan_book("9780140328721", "M001")
print(f"Loaned '{loan.book.title}' to {loan.member.name}")
print(f"Loan date: {loan.loan_date.strftime('%Y-%m-%d %H:%M')}")
print(f"Due date: {loan.due_date.strftime('%Y-%m-%d %H:%M')}")

# Check overdue
print(f"Is overdue: {loan.is_overdue()}")
print(f"Days overdue: {loan.days_overdue()}")

# Return book
print("\n=== Return Transaction ===")
library.return_book("9780140328721")
print(f"Book returned on: {loan.return_date.strftime('%Y-%m-%d %H:%M')}")
print(f"Is overdue: {loan.is_overdue()}")

# Member history
print("\n=== Member Loan History ===")
member_loans = library.get_member_loans("M001")
print(f"{member1.name} has {len(member_loans)} loan(s):")
for l in member_loans:
    status = "Returned" if l.return_date else "Active"
    print(f"  - {l.book.title} ({status})")

# Test validation
print("\n=== Testing Validation ===")
try:
    bad_book = Book("123", "Test", "Author", 2000)
except ValueError as e:
    print(f"Caught error: {e}")

try:
    bad_member = Member("M002", "Test", "invalid-email")
except ValueError as e:
    print(f"Caught error: {e}")

# Test duplicate loan
try:
    library.loan_book("9780451524935", "M001")
    library.loan_book("9780451524935", "M001")  # Should fail
except ValueError as e:
    print(f"Caught error: {e}")
````
