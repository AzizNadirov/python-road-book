````python
class Rectangle:  # noqa: F811
    def __init__(self, 
                 width: int|float, 
                 height: int|float):
        self.width = width
        self.height = height
        self._area = self.width * self.height

    @property
    def area(self)->float|int:
        """the area of the rectangle"""
        return self._area
    
rect1 = Rectangle(10, 20)
print(rect1.area)
````

4. `area.setter` ilÉ™ istifadÉ™Ã§inin dÃ¼zbucaqlÄ±nÄ±n sahÉ™sinÉ™ yanlÄ±ÅŸ É™dÉ™d mÉ™nimsÉ™tmÉ™sinin qarÅŸÄ±sÄ±nÄ± alÄ±n.

````python
class Rectangle:  # noqa: F811
    def __init__(self, 
                 width: int|float, 
                 height: int|float):
        self.width = width
        self.height = height
        self._area = self.width * self.height

    @property
    def area(self)->float|int:
        """the area of the rectangle"""
        return self._area
    
    @area.setter
    def area(self, value):
        if not isinstance(value, (int, float)) or value < 0:
            raise ValueError("Area must be a positive number!")
        self._area = value
    
rect1 = Rectangle(10, 20)   
print("Area: ", rect1.area) # Area:  200
rect1.area = -200   # ValueError: Area must be a positive number!
````

Ä°lk Ã¶ncÉ™, sadÉ™ `Currency` sinfi ataq, `__currencies` xÃ¼susiyyÉ™tindÉ™ `eur-data.json` faylÄ±ndan oxuduÄŸumuz verilÉ™nlÉ™ri saxlayaq:

````python
"""
    `get_currency_data` funksiyasÄ± `data/eur-data.json` faylÄ±ndakÄ± verilÉ™nlÉ™ri bizÉ™ bir lÃ¼ÄŸÉ™t ÅŸÉ™klindÉ™ qaytarÄ±r. AÃ§Ä±b baxsaz, gÃ¶rÉ™rsiz ki .json geniÅŸlÉ™nmÉ™li bu fayl elÉ™ lÃ¼ÄŸÉ™tÉ™ bÉ™nzÉ™yir. 
    Json faylÄ± oxumaq Ã¼Ã§Ã¼n Python standart kitabxanasÄ±na daxil olan `json` kitabxanasÄ±ndan istifadÉ™ edirik.
"""
import json


def get_currency_data() -> dict:
    """ load currency from file """
    data_path = "data/eur-data.json"
    with open(data_path, 'r') as f:
        data = json.load(f)
        return data
    
    
class Currency:
    """Simple Currency class"""
    def __init__(self,
                 code: str,
                 coeff: float):
        """
        Args:
            - code - currency code, e.g. AZN for Azerbaijan Manat
            - coeff - currency coefficient, e.g. 1 for 1 AZN, where `code = AZN'
        """
        self.code = code
        self.coeff = float(coeff)
        self.__currencies = get_currency_data() # `eur-data.json` verilÉ™nlÉ™rimiz

    def __str__(self)->str:
        return f"Currency: {self.code} [{self.coeff}]"
    
    def __repr__(self)->str:
        return self.__str__()
    

azn1 = Currency(code="AZN", coeff=1)
print(azn1) # Currency: AZN [1.0]
````

HazÄ±rda sadÉ™ - Ã¶zÃ¼ndÉ™ sadÉ™cÉ™ valyutanÄ±n adÄ±nÄ± vÉ™ miqdarÄ±nÄ± saxlaya bilÉ™n sinif yazdÄ±q. NÃ¶vbÉ™ti vÉ™ É™n É™sas mÉ™sÉ™lÉ™ - valyutaya bir Ã§É™ki, dÉ™yÉ™r vermÉ™kdir. BÃ¼tÃ¼n valyutalar Ã¼Ã§Ã¼n ortaq bir dÉ™yÉ™r olmalÄ±dÄ±r ki, bu ortaq dÉ™yÉ™rÉ™ gÃ¶rÉ™ valyutalarÄ± mÃ¼qayisÉ™ edÉ™, cÉ™mlÉ™yÉ™ vÉ™ sairÉ™ edÉ™ bilÉ™k. VerilÉ™n json faylÄ±na baxsaq(`data/eur-data.json`), gÃ¶rÉ™rik ki bÃ¼tÃ¼n valyutalar Ã¼Ã§Ã¼n dÉ™yÉ™rlÉ™r Avro(`EUR`) ilÉ™ verilib - `value` sahÉ™si. Bundan istifadÉ™ etmÉ™k olar. MÉ™sÉ™lÉ™n, rus rublu(`RUB`) Ã¼Ã§Ã¼n bu dÉ™yÉ™r bu dÉ™yÉ™r - 90.9299187002, bu demÉ™kdir ki `1 EUR = 90.9299187002 RUB`, Manat(`AZN`) Ã¼Ã§Ã¼n bu dÉ™yÉ™r: `1 EUR = 1.9275247084 AZN`. BelÉ™liklÉ™, hÉ™r bir valyuta Ã¼Ã§Ã¼n bÉ™lli olan avro dÉ™yÉ™rindÉ™n istifadÉ™ edÉ™cÉ™yik. 

Ä°lk addÄ±m kimi, `Currency` sinfinÉ™ `_unit_w` adlÄ± xÃ¼susiyyÉ™t É™lavÉ™ edÉ™k. Bu xÃ¼susiyyÉ™t verilÉ™n valyutanÄ±n 1 É™msalÄ± Ã¼Ã§Ã¼n olan dÉ™yÉ™ri olacaq vÉ™ belÉ™ hesablanacaq: 

`self._unit_w = 1 / self.__currencies [self.code]['value']`

Bu minvalla, `AZN` Ã¼Ã§Ã¼n bu dÉ™yÉ™r: 0.5188000940491602, `RUB` Ã¼Ã§Ã¼n: 0.010997480414527198, `EUR` Ã¼Ã§Ã¼nsÉ™: 1.0 olacaq.

`_unit_w` - valyutanÄ±n 1 É™msalÄ± Ã¼Ã§Ã¼n olan dÉ™yÉ™ri buildirir, mÉ™sÉ™lÉ™n, 1 AZN, 3 RUB vÉ™ sairÉ™. BizdÉ™ isÉ™, É™lavÉ™ olaraq `coeff` sahÉ™si dÉ™ var. MÉ™sÉ™lÉ™n, `Currency(code="AZN", coeff=5)` Ã¶zÃ¼ndÉ™n "5 manat" ehtiva edir. Ona gÃ¶rÉ™ dÉ™, Ã¶zÃ¼ndÉ™n currency nÃ¼sxÉ™sinin Ã¼mumi dÉ™yÉ™rini ehtiva edÉ™cÉ™k `self._w = self._unit_w * self.coeff` sahÉ™si É™lavÉ™ edÉ™k. `_w` - nÃ¼sxÉ™nin Ã¼mumi Ã§É™kisini, dÉ™yÉ™rini bildirÉ™cÉ™k vÉ™ gÃ¶rÃ¼ndÃ¼yÃ¼ kimi - `vahidin dÉ™yÉ™ri * É™msal` kimi hesablanÄ±r. 



````python
class Currency:
    """ currency class """
    def __init__(self, 
                 code: str, 
                 coeff: float|int):
        """ create currency """
        self.__currencies = get_currency_data()
        if code not in self.__currencies .keys():
            raise ValueError(f"Currency code not found! Known ones: \n\t{list(self.__currencies.keys())}")
        
        self.code = code
        self.coeff = coeff
        self._unit_w = 1 / self.__currencies [self.code]['value']   # vahid dÉ™yÉ™ri
        self._w = self._unit_w * self.coeff     # Ã¼mumi dÉ™yÉ™ri
        
    
    def __str__(self)->str:
        return f"Currency: {self.code} [{self.coeff}]"
    
    def __repr__(self)->str:
        return self.__str__()
    
    
    def __add__(self, other: 'Currency')->'Currency':
        # ancaq vÉ™ ancaq baÅŸqa bir `Currency` nÃ¼sxÉ™si ilÉ™ cÉ™mlÉ™yÉ™ bilÉ™rik
        if not isinstance(other, Currency):
            raise ValueError(f'Can add only with other Currency, not {type(other)} object')
        # toplamanÄ±n nÉ™ticÉ™si - cÉ™m dÉ™ `Currency` nÃ¼sxÉ™si olacaq.
        # cÉ™min parametrlÉ™rini hesablayaq
        res_w = self._w + other._w
        res_coef = res_w / self._unit_w
        
        return Currency(
                    code=self.code,
                    coeff=res_coef)

    
azn1 = Currency(code="AZN", coeff=1)
rb_100 = Currency(code="RUB", coeff=100)

# 1 AZN + 1 AZN = 2 AZN
print(azn1 + azn1)  # Currency: AZN [2.0]


# 1 AZN + 100 RUB = 3.11 AZN
print(azn1 + rb_100)  # Currency: AZN [3.119791522914625]
````

````python
import json


def get_currency_data() -> dict:
    """ load currency from file """
    data_path = "/home/tengo/Documents/book_cs_with_python/code/11_1/my_calc_app/data/eur-data.json"
    with open(data_path, 'r') as f:
        data = json.load(f)
        return data
    

class Currency:
    """ currency class """
    def __init__(self, 
                 code: str, 
                 coeff: float|int):
        """ create currency """
        self.__currencies = get_currency_data()
        if code not in self.__currencies .keys():
            raise ValueError(f"Currency code not found! Known ones: \n\t{list(self.__currencies.keys())}")
        
        self.code = code
        self.coeff = coeff
        self._unit_w = 1 / self.__currencies [self.code]['value']
        self._w = self._unit_w * self.coeff
        
    
    def __str__(self)->str:
        return f"Currency: {self.code} [{self.coeff}]"
    
    def __repr__(self)->str:
        return self.__str__()
    
    
    def __add__(self, other: 'Currency')->'Currency':
        if not isinstance(other, Currency):
            raise ValueError(f'Can add only with other Currency, not {type(other)} object')
        
        res_w = self._w + other._w
        res_coef = res_w / self._unit_w
        
        return Currency(
                    code=self.code,
                    coeff=res_coef)
        
    def __sub__(self, other: 'Currency')->'Currency':
        if not isinstance(other, Currency):
            raise ValueError(f'Can subtract only with other Currency, not {type(other)} object')
        
        res_w = self._w - other._w
        res_coef = res_w / self._unit_w
        
        return Currency(
                    code=self.code,
                    coeff=res_coef)
        
        
    def __mul__(self, other: int|float)->'Currency':
        if not isinstance(other, (int, float)):
            raise ValueError(f'Can multiplicate only with numbers, not {type(other)} object')
        
        res_w = self._w * other
        res_coef = res_w / self._unit_w
        
        return Currency(
                    code=self.code,
                    coeff=res_coef)
        
    
    def __rmul__(self, other: int|float)->'Currency':
        if not isinstance(other, (int, float)):
            raise ValueError(f'Can multiplicate only with numbers, not {type(other)} object')
        
        res_w = self._w * other
        res_coef = res_w / self._unit_w
        
        return Currency(
                    code=self.code,
                    coeff=res_coef)
        
        
    def __truediv__(self, other: int|float)->'Currency':
        if not isinstance(other, (int, float)):
            raise ValueError(f'Can multiplicate only with numbers, not {type(other)} object')
        
        res_w = self._w / other
        res_coef = res_w / self._unit_w
        
        return Currency(
                    code=self.code,
                    coeff=res_coef)
        
    
    def __rtruediv__(self, other: int|float)->'Currency':
        if not isinstance(other, (int, float)):
            raise ValueError(f'Can multiplicate only with numbers, not {type(other)} object')
        
        res_w = self._w / other
        res_coef = res_w / self._unit_w
        
        return Currency(
                    code=self.code,
                    coeff=res_coef)
        
    
    def __rshift__(self, to: str)->'Currency':
        """ convert to other currency """
        if to not in self.__currencies.keys():
            raise ValueError(f"Unknown currency: {to}")
        
        to_unit_w = self.__currencies.get(to)['value']
        coeff = self._w * to_unit_w
        
        return Currency(code=to, coeff=coeff)
````
