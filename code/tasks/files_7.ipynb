````python
users_dict = [
    {"name": "Rəşad", "email": "reshad@school.edu", "pass": "12345678pass"}, 
    {"name": "Əziz", "email": "eziz@school.edu", "pass": "pass"}, 
    {"name": "Ülvi", "email": "ulvi@school.edu", "pass": "123pass"}, 
    {"name": "Hikmət", "email": "hikmet@school.edu"},
]

def write_students(students: dict[str, str], 
                   file_name: str="students.csv")->None:
    """ write students into file """
    null_value = ""
    with open(file_name, 'a') as file:
        print(f":Writing {len(students)} students to {file_name}")
        for student in students:
            print(f":::Writing student: {student}")
            name = student.get("name", null_value)
            email = student.get("email", null_value)
            password = student.get("password", null_value)
            file.write(f"{name},{email},{password}\n")

write_students(users_dict, file_name)
````
