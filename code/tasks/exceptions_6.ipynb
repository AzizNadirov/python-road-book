1. Təhlükəsiz şəkildə sətrləri ədədlərə çevirəcək `str_to_num` funksiyası yazın. Funksiya ötürülmüş sətri ədədə çevirməlidir. Əgər ədədə çevrilmə zamanı xəta baş verərsə, geriyə `None` qaytarılsın. 
```python
def str_to_num(string_num: str)-> float|None:
    pass

str_to_num("3.1415")   # 3.1415
str_to_num("-3")   # -3
str_to_num("3a")   # None
```

````python
def str_to_num(string_num: str)->float|None:
    """ convert string to number """
    try:
        return float(string_num)
    except ValueError:
        return None


print(str_to_num('3.1415'))
print(str_to_num('-3'))
print(str_to_num('3a'))
````

2. Düzbucaqlının sahəsini hesablayacaq `rectangle_area` funksiyasını yazın. Funksiya - dördbucaqlının eni və uzunluğunu ifadə edən `length` və `width` parametrləri almalıdır. Sadə `assert` yoxlamaları ilə əmin olun ki bu parametrlər müsbət ədədlərdir.

```python
def rectangle_area(length: float, width: float)->float:
    # Your implementation here
```

````python
def rectangle_area(length: float, width: float)->float:
    assert length > 0 and width > 0, 'length and width must be positive numbers'
    return length * width
````

həmçinin, `isinstance(nüsxə, class_obyekti)` - funksiyası ilə, əvvəlcə giriş parametrlərinin tiplərini yoxlaya bilərdik,
əgər ədədlərdən fərqli bir şeylər olarsa(məsələn, sətr ötürülə bilərdi), o zaman bir xəta ata bilərik.

````python
isinstance(3, int)  # True
isinstance(3.0, float)  # False
isinstance(3, (int, float))  # True, siniflər yazısı ötürmək olar.

def rectangle_area_v2(length: float, 
                      width: float)->float:
    if not (isinstance(length, (int, float)) and isinstance(width, (float, int))):
        raise ValueError("All rectangle sizes must be numbers!")
    
    assert length > 0 and width > 0, 'length and width must be positive numbers'
    return length * width


rectangle_area_v2(3, '5')   # ValueError: All rectangle sizes must be numbers!
````

3. Istifadəçi parolunu yoxlamaq - validasiya etmək üçün `validate_password` funksiyasını yazın. Funksiya özündən parol ehtiva edən sətr tipli `password` arqumentini alır və aşağıdakı şərtlərdən ən azı biri ödənməsə "Invalid password!" ismarıclı `ValueError` xətası atmalıdır, hər şey qaydasındadırsa - `True` qaytarmalıdır.
    - Ən azı 8 simvol olmalır;
    - boşluq simvolları olmamalıdır;
    - İçərisində həm hərflər, həm də rəqəmlər olmalıdır;

```python
def validate_password(password: str)->bool:
    pass

validate_password("12345678pass")  # True
validate_password("12345678")  # ValueError
validate_password(" 12345678pass")  # ValueError
```

````python
def validate_password(password: str)->bool:
    """ raise error if password is not valid """
    # 0: type
    if not isinstance(password, str):
        raise TypeError("Password must be a string!")
    # 1: length
    if len(password) < 8:
        raise ValueError("Password must be at least 8 characters long!")
    # 2: whitespace
    if any(char.isspace() for char in password):
        raise ValueError("Password must not contain whitespace!")
    # 3: letters and digits
    if not (any(char.isalpha() for char in password) and any(char.isdigit() for char in password)):
        raise ValueError("Password must contain at least one letter and one digit!")
    
    return True

validate_password("12345678pass")  # True
````

````python
validate_password('12345678')   # ValueError: Password must contain at least one letter and one digit!
````

````python
validate_password(" 12345678pass")  # ValueError: Password must not contain whitespace!
````

4. Yuxarıdakı `validate_password` funksiyasını elə dəyişin ki parol yoxlamanı keçmədiyi təqdirdə xəta atmaq əvəzinə geriyə `False` qaytarılsın. Təsəvvür edin ki belə bir istifadəçilər bazamız var:
```python
users_dict = [
    {"name": "Rəşad", "email": "reshad@school.edu", "pass": "12345678pass"}, 
    {"name": "Əziz", "email": "eziz@school.edu", "pass": "pass"}, 
    {"name": "Ülvi", "email": "ulvi@school.edu", "pass": "123pass"}, 
    {"name": "Hikmət", "email": "hikmet@school.edu"},
]
```
Istifadəçilərin parollarını yoxlamaq üçün `validate_users_pass` adlı funksiya yazın. Funksiya bu arqumentləri almalıdır:
- `users: dict` - yuxarıdakı nümunəyə bənzər istifadəçilər lüğəti
- `pass_field_name: str="pass"` - lüğətdə özündə istifadəçi parolunu saxlamalı olan açar, susmaya görə `"pass"` olacaq.

`validate_users_pass` funksiyası bütün istifadəçilərin parollarını yoxladıqdan sonra geriyə lüğət qaytarmalıdır: lüğətin açarları - yoxlanılan istifadəçinin(siyahıdakı lüğətin) indeksi, dəyərləri isə  `validate_password` funksiyasının bu istifadəçinin parolu üçün nəticəsi olacaq. Onu da nəzərə alın ki, bəzi istifadəçilərin parolları olmaya da bilər. Belə olan halda, həmin istifadəçi üçün nəticə `None` olacaq.
```python
def validate_password(password: str)->bool:
    pass

def validate_users_pass(users: dict, 
                        pass_field_name: str="pass")-> dict[int, bool]
    pass

res = validate_users_pass(users_dict)
print(res)  # {0: True, 1: False, 2: False, 3: None}
```

````python
# 1. İlk öncə, deyildiyi kimi - funksiyamızı dəyişək:

def validate_password(password: str)->bool:
    """ raise error if password is not valid """
    # 0: type
    if not isinstance(password, str):
        return False
    
    # 1: length
    if len(password) < 8:
        return False
    
    # 2: whitespace
    if any(char.isspace() for char in password):
        return False
    
    # 3: letters and digits
    if not (any(char.isalpha() for char in password) and any(char.isdigit() for char in password)):
        return False
    
    return True


## Bir sətrə yazma həvəskarları bu funksiyanı bir sətrə də yaza bilər:
def validate_password(password: str) -> bool:
    return isinstance(password, str) and len(password) >= 8 and not any(char.isspace() for char in password) and any(char.isalpha() for char in password) and any(char.isdigit() for char in password)
````

````python
validate_password("12345678pass\n")  # False
validate_password("12345678pass")  # True
````

````python
# ŞƏrtdə verilmiş istifadəçilər lüğəti
# göründüyü kimi, parollar `pass` açarı ilədir. Üstəlik, sonuncu istifadəçinin parolu yoxdur.
users_dict = [
    {"name": "Rəşad", "email": "reshad@school.edu", "pass": "12345678pass"}, 
    {"name": "Əziz", "email": "eziz@school.edu", "pass": "pass"}, 
    {"name": "Ülvi", "email": "ulvi@school.edu", "pass": "123pass"}, 
    {"name": "Hikmət", "email": "hikmet@school.edu"},
]
````

````python
def validate_users_pass(users: list[dict], 
                        pass_field_name: str="pass")-> dict[int, bool]:
    validated_users = {}
    for user in users:
        # istifadəçinin parolu
        user_password = user.get(pass_field_name, None)
        # əgər parol yoxdursa None kimi qeyd et
        if user_password is None:
            validated_users[users.index(user)] = None
        else:   # əks halda, parolu yoxlayıb nəticəsini qeyd et
            validated_users[users.index(user)] = validate_password(user_password)
    return validated_users


res = validate_users_pass(users_dict)
print(res)  # {0: True, 1: False, 2: False, 3: None}
````

Funksiyamızı daha da qısa yaza bilərik:

````python
def validate_users_pass(users: list[dict], 
                        pass_field_name: str="pass")-> dict[int, bool]:
    return {users.index(user): validate_password(user.get(pass_field_name, None)) if user.get(pass_field_name, None) is not None else None for user in users}


res = validate_users_pass(users_dict)
print(res)  # {0: True, 1: False, 2: False, 3: None}
````
